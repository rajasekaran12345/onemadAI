```python
import pandas as pd
import numpy as np


data_dict = {
    'model_engine_number': [101, 102, 103, 104, 105, 101, 101, 103, 105, 104, 104],
    'units_sold': [10, 15, 12, 18, 20, 11, 13, 14, 19, 17, 16],
    'date_sold': ['2023-10-26 ', '2023-10-26 ', '2023-10-26 ', '2023-10-27 ', '2023-10-27 ', '2023-10-28', '2023-10-28', '2023-10-28', '2023-10-29', '2023-10-29', '2023-10-29'],
    'demand_category': ['High', 'Low', 'High', 'High', 'Low', 'High', 'High', 'Low', 'Low', 'High', 'High'],
    'price': [1500, 1200, 1600, 1800, 2000, 1550, 1520, 1250, 1900, 1750, 1780]
}
df = pd.DataFrame(data_dict)


num_new_rows = 150 - len(df)
new_data = {}
for col in df.columns:
    new_data[col] = np.tile(df[col], num_new_rows // len(df) + 1)[:num_new_rows]

new_df = pd.DataFrame(new_data)


df_expanded = pd.concat([df, new_df], ignore_index=True)


df_expanded['price'] = df_expanded['units_sold'] * 100 + np.random.randint(-50, 50, size=len(df_expanded))

df_expanded = df_expanded.sample(frac=1).reset_index(drop=True)


```

creating a model for predicting demand.

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=42)
ss_demand=StandardScaler()
x2_train_scaled=ss_demand.fit_transform(x2_train)
x2_test_scaled=ss_demand.transform(x2_test)
```

```python
y2_train = y2_train.values.ravel()
y2_test = y2_test.values.ravel()
'''transforms into 1d array'''
```

```python
from sklearn.linear_model import LogisticRegression
model_demand=LogisticRegression()
model_demand.fit(x2_train_scaled,y2_train)
model_demand.score(x2_test_scaled,y2_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rf=RandomForestClassifier()
param_grid_demand = {'n_estimators': [10, 50, 100]}
grid_search_demand = GridSearchCV(estimator=rf,param_grid=param_grid_demand, cv=5, scoring='accuracy')
grid_search_demand.fit(x2_train_scaled, y2_train)
```

```python
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x2_train_scaled,y2_train)
rf.score(x2_test_scaled,y2_test)
```

predicting the price and demand for the engine_model using users input

```python
def predict_price_demand():
  import datetime
  import numpy as np
  model_engine_number= int(input("ENTER THE MODEL ENGINE NUMBER: "))
  units_sold= int(input("ENTER THE NUMBER OF UNITS SOLD"))
  day= int(input("ENTER THE DATE(DAY)"))
  month= int(input("ENTER THE MONTH"))
  year= int(input(("ENTER THE YEAR")))

  input_array= np.array([[model_engine_number,units_sold,day,month,year]]).reshape(1,-1)
  input_scaled=ss_price.transform(input_array)
  price=model_price.predict(input_scaled)
  print("THE PRICE OF THE ENGINE OF THE MODEL_NO",model_engine_number,"AT THE DATE OF",datetime.date(year, month, day),"is",price)
  input_demand = np.array([[model_engine_number, units_sold, price[0], day, month, year]]).reshape(1, -1)
  input_demand_scaled=ss_demand.transform(input_demand)
  demand_output=model_demand.predict(input_demand_scaled)
  if demand_output==0:
    print("THE DEMAND OF THE ENGINE OF THE MODEL_NO",model_engine_number,"IS VERY HIGH")
  else:
    print("THE DEMAND OF THE ENGINE OF THE MODEL_NO",model_engine_number,"IS VERY LOW")
```
