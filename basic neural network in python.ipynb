```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
data=pd.read_csv("diabetes.csv")
```

```python
data.head(5)
```

```python
predictors=data.iloc[:,0:8]
response=data.iloc[:,8]
```

```python
x_train,x_test,y_train,y_test=train_test_split(predictors,response,test_size=0.2)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
kerasmodel=Sequential()#initializing the model
kerasmodel.add(Dense(12,input_dim=8,activation='relu'))
kerasmodel.add(Dense(8,activation='relu'))#using relu function to avoid the exploiting and vanishing gradient problem
kerasmodel.add(Dense(1,activation='sigmoid'))#hence it has a output value of 0 and 1 we use sigmoid function
              
```

```python
#compiling model
kerasmodel.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
#fitting model
kerasmodel.fit(x_train,y_train,epochs=150,batch_size=10)
```

```python
#train accuracy 
_,accuracy=kerasmodel.evaluate(x_train,y_train)
print("the accuracy of train score is {0}".format(accuracy*100))
```
