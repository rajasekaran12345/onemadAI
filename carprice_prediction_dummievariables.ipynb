```python
#importing pandas and matplotlib
import pandas as pd
from matplotlib import pyplot
```

```python
#converting into dataframe
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
#changing the columns
df.rename(columns={'Car Model':'cm'},inplace='true')
```

```python
df
```

```python
#converting categorical data into binary form,using dummie_variables
sep=pd.get_dummies(df.cm)
```

```python
sep
```

```python
merged=pd.concat([df,sep],axis='columns')
```

```python
merged.shape
```

```python
merged
```

```python
final=merged.drop("cm",axis='columns')
```

```python
final
```

```python
#changing columns name
final.rename(columns={'Sell Price($)':'sp$','Audi A5':'audi','BMW X5':'bmw','Mercedez Benz C class':'benz'},inplace=True)
```

```python
final
```

```python
#dropping the benz column,because of multi_colinearity
finale=final.drop('benz',axis='columns')
```

```python
finale
```

```python
x=finale.drop('sp$',axis='columns')
```

```python
x
```

```python
y=finale['sp$']
```

```python
y
```

```python
#importing linear regression
from sklearn.linear_model import LinearRegression
```

```python
#initailizing linear_regression model
model=LinearRegression()
```

```python
#training our model usinf .fit method
model.fit(x,y)
```

```python
#testing accuracy of the model
model.score(x,y)
```

```python
#predicting our model
z=model.predict(x)
z
```
